```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. load data dulu
df = pd.read_csv("BlackmoresTokopedia.csv")

# 2. baru visualisasi
plt.figure(figsize=(8,5))
sns.histplot(df["price_int"], bins=30, kde=False)
plt.xlabel("Harga produk (rupiah)")
plt.ylabel("Jumlah listing")
plt.title("Histogram distribusi harga produk Blackmores di Tokopedia")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["stock"], bins=30, kde=False)
plt.xlabel("Jumlah stok")
plt.ylabel("Jumlah listing")
plt.title("Histogram distribusi stok produk Blackmores")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("BlackmoresTokopedia.csv")

print(df.columns)  # cek dulu

# 1) rename dulu
df = df.rename(columns={"shop.city": "shop_city"})

print(df.columns)  # pastikan sekarang ada 'shop_city'

# 2) baru groupby
city_counts = df.groupby("shop_city")["name"].count().sort_values(ascending=False)

plt.figure(figsize=(10,5))
city_counts.plot(kind="bar")
plt.xlabel("Kota")
plt.ylabel("Jumlah listing")
plt.title("Jumlah listing produk Blackmores per kota")
plt.tight_layout()
plt.show()
```

```python
# cek dulu kolom
print(df.columns)

# rename kolom- kolom yang pakai titik
df = df.rename(columns={
    "shop.city": "shop_city",
    "shop.is_official": "shop_is_official"
})

print(df.columns)  # pastikan sekarang ada shop_city dan shop_is_official

# contoh pie chart status toko
status_counts = df["shop_is_official"].value_counts()

plt.figure(figsize=(6,6))
status_counts.plot(kind="pie", autopct="%1.1f%%", labels=["Non-official","Official"])
plt.ylabel("")
plt.title("Proporsi status toko penjual Blackmores")
plt.tight_layout()
plt.show()
```

```python
# 20 produk dengan listing terbanyak (pakai kolom 'name')
product_counts = df["name"].value_counts().head(20)

plt.figure(figsize=(10,5))
product_counts.plot(kind="bar")
plt.xlabel("Nama produk")
plt.ylabel("Jumlah listing")
plt.title("20 produk Blackmores dengan listing terbanyak")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("BlackmoresTokopedia.csv")

price_desc  = df["price_int"].describe()
price_mean  = df["price_int"].mean()
price_median = df["price_int"].median()
price_mode   = df["price_int"].mode()[0]

print(price_desc)
print("Mean  :", price_mean)
print("Median:", price_median)
print("Mode  :", price_mode)
```

```python
stock_desc  = df["stock"].describe()
stock_mean  = df["stock"].mean()
stock_median = df["stock"].median()
stock_mode   = df["stock"].mode()[0]

print(stock_desc)
print("Mean  :", stock_mean)
print("Median:", stock_median)
print("Mode  :", stock_mode)
```
